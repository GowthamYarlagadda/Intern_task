# Function to extract text from a PDF file

```python

import PyPDF2


def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, "rb") as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)

        # Iterate through the pages and extract text
        for page in pdf_reader.pages:
            text += page.extract_text()

    return text


# Example usage:
pdf_path = "/content/resume-dataset/data/data/TEACHER/28642819.pdf"
cv_text = extract_text_from_pdf(pdf_path)
print(cv_text)
```

# Function to extract Key-details from a PDF file

```python
import spacy
import re

# Load the English language model
nlp = spacy.load("en_core_web_sm")

def extract_key_details(cv_text):
    # Define keyword patterns for section headings
    section_patterns = {
        "Category": re.compile(r"(?i)category\s*:?|job\s*objective"),
        "Skills": re.compile(r"(?i)core\s*strengths|summary\s*of\s*qualifications|skills"),
        "Education": re.compile(r"(?i)educational\s*background|education"),
        "Work Experience": re.compile(r"(?i)work\s*experience"),
        "Certifications": re.compile(r"(?i)certifications")
    }

    # Initialize a list to store extracted details
    details = []

    # Initialize a variable to keep track of the current section
    current_section = None

    # Initialize a list to store lines within the current section
    current_section_lines = []

    # Split the text into lines and iterate through them
    lines = cv_text.splitlines()
    for line in lines:
        # Check if the line matches any section heading pattern
        for section, pattern in section_patterns.items():
            if pattern.search(line):
                # If a new section is encountered, add the previous section's lines to details
                if current_section is not None and current_section_lines:
                    details.append([current_section] + current_section_lines)
                current_section = section
                current_section_lines = []
                break

        # If the line doesn't match a section heading and a current section is identified, add it to the current section's lines
        if current_section is not None and line.strip():
            current_section_lines.append(line.strip())

    # Add the last section's lines to details
    if current_section is not None and current_section_lines:
        details.append([current_section] + current_section_lines)

    return details


details = extract_key_details(cv_text)

# Print the extracted details as a list of lists
for detail in details:
    section = detail[0]
    content_lines = detail[1:]
    print(section)
    for line in content_lines:
        print(line)
```

# This code run on GPU and it's faster the above code

```python
from transformers import DistilBertTokenizer, DistilBertModel
from sklearn.metrics.pairwise import cosine_similarity
from tqdm import tqdm
import numpy as np
import torch

# Check if a GPU is available and set the device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Initialize DistilBERT tokenizer and model on the selected device
tokenizer = DistilBertTokenizer.from_pretrained("distilbert-base-uncased")
model = DistilBertModel.from_pretrained("distilbert-base-uncased").to(device)

# Initialize a dictionary to store results
matched_cvs = {}
flattened_cvs = [(cv_content, cv_file) for cv_file, cv_content in cvs]

# Define batch size
batch_size = 32

# Iterate through each job description with tqdm
for idx, job_description in tqdm(enumerate(job_descriptions, start=1), desc="Matching Job Descriptions"):
    print("\n")
    print(f"Job Description {idx}:")

    # Tokenize and convert to embeddings for the job description on the selected device
    job_desc_tokens = tokenizer(job_description, padding=True, truncation=True, return_tensors="pt").to(device)
    job_desc_embeddings = model(**job_desc_tokens).last_hidden_state.mean(dim=1).detach().cpu().numpy()

    cv_scores = []

    # Iterate through each CV in batches
    for batch_start in range(0, len(flattened_cvs), batch_size):
        batch_end = min(batch_start + batch_size, len(flattened_cvs))
        batch_cv_texts = [cv_text for cv_text, _ in flattened_cvs[batch_start:batch_end]]

        # Tokenize and convert to embeddings for the batch of CVs on the selected device
        batch_cv_tokens = tokenizer(batch_cv_texts, padding=True, truncation=True, return_tensors="pt").to(device)
        batch_cv_embeddings = model(**batch_cv_tokens).last_hidden_state.mean(dim=1).detach().cpu().numpy()

        # Calculate cosine similarity for the batch
        batch_similarity_scores = cosine_similarity(job_desc_embeddings, batch_cv_embeddings)

        # Append the batch results to cv_scores
        for i, (cv_text, cv_file_path) in enumerate(flattened_cvs[batch_start:batch_end]):
            similarity_score = batch_similarity_scores[0][i]
            cv_scores.append((cv_text, cv_file_path, similarity_score))

    # Sort and select the top 5 CVs
    top_5_cvs = sorted(cv_scores, key=lambda x: x[2], reverse=True)[:5]
    matched_cvs[f"Job Description {idx}"] = top_5_cvs

# Now you can print the matched CVs for each job description
for job_desc, top_cvs in matched_cvs.items():
    # print(job_desc)
    print(f"{job_desc} - Top 5 Matched CVs:")
    for cv_text, cv_file_path, score in top_cvs:
        print(f"Similarity Score: {score:.4f}")
        print(f"CV File Path: {cv_file_path}")
        print("Common Points: " + cv_text)
        print("\n")
```
